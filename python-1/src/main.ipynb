**Проблема 1**

Виправлю описку в назві стовпця 

````python
df.columns = ['Country Name', 'Region', 'GDP per capita', 'Population', 'CO2 emission', 'Area']
df.columns
````

**Проблема 2**

Використаю регулярні вирази для того, щоб замінити коми на крапки. Також приведу дані до типу float. 

````python
cols_to_update = ['GDP per capita', 'CO2 emission', 'Area']

df_correct = df.copy()
df_correct[cols_to_update] = df_correct[cols_to_update].apply(lambda x: pd.to_numeric(x.str.replace(',', '.', regex=True), errors='coerce'))
df_correct
````

**Проблема 2**

Виправлю значення від'ємних даних на додатні, за допомогою команди `abs()`

````python
df_correct['Area'] = df_correct['Area'].abs()
df_correct['Area'].sort_values()
````

````python
df_correct['GDP per capita'] = df_correct['GDP per capita'].abs()
df_correct['GDP per capita'].sort_values()
````
